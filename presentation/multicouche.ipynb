```python
from source.monocouche_matrice import importSigne, signe, predict

trainingDataset, validationDataset, indexesValidation, row_num, output_num = importSigne('Datas\\LangageDesSignes\\data_formatted.csv')
monocouche = signe(trainingDataset, row_num, output_num, maxIteration=10000)
```

```python
import numpy as np

# Predict results for all elements in validationDataset[0]
predicted_results = np.array([predict(monocouche, element) for element in validationDataset[0]])

# Assume predicted_results is a 2D numpy array
max_values = np.max(predicted_results, axis=1)
max_indices = np.argmax(predicted_results, axis=1)
max_indices += 1

classes = ["class" + str(index) for index in max_indices]

# Combine max_values and max_indices into a 2D array
prediction = np.array(list(zip(classes, max_values, indexesValidation.ravel())))


np.savetxt("output/prediction.csv", prediction, delimiter=",", fmt="%s")
np.savetxt("output/validation.csv", np.transpose(validationDataset[1]), delimiter=",", fmt="%s")
```
